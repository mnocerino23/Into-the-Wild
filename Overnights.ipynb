# Investigating the lack of sign ups during the 2021-2022 school year



Introduce a new feature titled 'Success'. If 85% of a trip's available tickets sell, we consider the trip a success. If the number is less, we consider the trip not succesfully filled. This will be a better metric than the 'Filled?' feature we currently have since it allows for trips with very few empty spots to be still considered succesful (how the org views them).

```python
overnight_trips['Success'] = overnight_trips['PercentSold'].apply(lambda x: 1 if x >= 0.85 else 0)
```

```python
overnight_trips.to_csv('Overnights.csv', index = False)
```

One-hot encode Trip Type, Skill Level, Terrain Type, Day Leave

```python
dummy_triptype = pd.get_dummies(overnight_trips['Trip Type'])
dummy_skill = pd.get_dummies(overnight_trips['Skill Level'])
dummy_terrain = pd.get_dummies(overnight_trips['Terrain Type'])
dummy_dayleave = pd.get_dummies(overnight_trips['Day Leave'])
dummy_quarter = pd.get_dummies(overnight_trips['Quarter'])

one_hot_encoded = [dummy_triptype, dummy_skill, dummy_terrain, dummy_dayleave, dummy_quarter]
for features in one_hot_encoded:
  overnight_trips = pd.merge(left = overnight_trips, right = features, left_index = True, right_index = True)

overnight_trips = overnight_trips.drop(columns = ['Trip Type', 'Skill Level', 'Terrain Type', 'Day Leave'])
overnight_trips.head(5)
```

We don't have large enough of a dataset to build classifiers or regression models for trip sign ups, but we can still use sklearn's selectkbest method to find the most import features 

```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
overnight_trips.columns
```

```python
features = overnight_trips.drop(['Year', 'Trip Name', 'Region', 'County', 'Day Return', 'Tickets Sold', 'Trip', 'Quarter', 'Tickets Available',
                                 'Empty spots', 'Filled?', 'PercentSold', 'Success'], axis = 1)
target = overnight_trips['Success']
```

```python
best_features = SelectKBest(f_classif, k=5)
k_best_data = best_features.fit_transform(features, target)

mask = k_best_data.get_support()
new_features = features.columns[mask]
```
