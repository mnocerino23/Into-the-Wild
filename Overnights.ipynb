Read in dataset containing information and sign up data for all of Into the Wild's overnight camping/backpacking trips during the 2021-2022 school year at Santa Clara University.

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
overnight_trips = pd.read_csv('/content/drive/MyDrive/Data_Science_Projects/Into_the_Wild/Into the Wild Trip Results 2021 - Overnights.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
print(overnight_trips.columns)
```

```python
#Add in feature that represents the  

overnight_trips = overnight_trips.assign(PercentSold=lambda x: (x['Tickets Sold'] / x['Tickets Available']))
```

# Recap of Into the Wild 2021 Trips - Trip Type, Skill Level, and Terrain

```python
sns.histplot(data=overnight_trips, x="Trip Type", color = "orange")
plt.title('2021-2022 Overnight Trip Types', fontsize=14)
```

```python
sns.histplot(data = overnight_trips, x = "Skill Level", color = "blue")
plt.title('2021-2022 Overnight Trip Levels', fontsize=14)
```

```python
sns.histplot(data = overnight_trips, x = "Terrain Type", color = "magenta")
plt.title('2021-2022 Overnight Trip Terrain', fontsize=14)
```

```python
sns.histplot(data = overnight_trips, x = "Day Leave")
plt.title('2021-2022 Overnight Trips Day of Departure', fontsize=14)
```

```python
sns.histplot(data = overnight_trips, x = "Number of Nights", color = 'yellow', bins = 2)
plt.title('2021-2022 Overnight Trips Day of Departure', fontsize=14)
```

```python
sns.histplot(data = overnight_trips, x = "Quarter", color = "brown")
plt.title('2021-2022 Overnight Trips by Quarter', fontsize=14)
```

# Where we got Wild:

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.histplot(data = overnight_trips, x = "Region",  bins = 20)
plt.title('2021-2022 Overnight Trips by Region', fontsize=25)
```

```python
sns.histplot(data = overnight_trips, x = "County",  bins = 20)
plt.title('2021-2022 Overnight Trips by County', fontsize=25)
```

# Trip Distances

```python
sns.histplot(data = overnight_trips, x = "Distance (miles)", bins = 5, kde = True, color = 'Green')
plt.title('Driving Distances of 2021-2022 ITW Overnights', fontsize=16)
```

Violin Plot using Seaborn:

```python
plt.title('Distribution of Distances of 2021-2022 ITW Overnights', fontsize=20)
sns.set_theme(style="whitegrid")
ax = sns.violinplot(x=overnight_trips["Distance (miles)"])
```

# Pricing of 2021-2022 Overnights

```python
sns.histplot(data = overnight_trips, x = "Price", bins = 10, kde = True, color = 'Purple', hue = 'Number of Nights')
plt.title('Histogram of 2021-2022 ITW Overnight Prices', fontsize=20)
plt.xlabel('Trip Price (USD)')
```

Violin Plot using seaborn:

```python
plt.title('Price Distribution of 2021-2022 ITW Overnights', fontsize=20)
sns.set_theme(style="whitegrid")
ax = sns.violinplot(x=overnight_trips["Price"])
```

# Investigating the lack of sign ups during the 2021-2022 school year

```python
sns.histplot(data = overnight_trips, x = "PercentSold", bins = 5, kde = True)
plt.title('Percent Full 2021-2022 Overnights', fontsize=20)
plt.xlabel('Trip Price (USD)')
```
