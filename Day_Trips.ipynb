# Investigating the lack of sign ups during the 2021-2022 school year

Introduce a new feature titled 'Success'. If 85% of a trip's available tickets sell, we consider the trip a success. If the number is less, we consider the trip not succesfully filled. This will be a better metric than the 'Filled?' feature we currently have since it allows for trips with very few empty spots to be still considered succesful (how the org views them).

```python
day_hikes['Success'] = day_hikes['PercentSold'].apply(lambda x: 1 if x >= 0.85 else 0)
```

```python
dummy_triptype = pd.get_dummies(day_hikes['Trip Type'])
dummy_skill = pd.get_dummies(day_hikes['Skill Level'])
dummy_terrain = pd.get_dummies(day_hikes['Terrain Type'])
dummy_dayleave = pd.get_dummies(day_hikes['Day of week'])
dummy_quarter = pd.get_dummies(day_hikes['Quarter'])

one_hot_encoded = [dummy_triptype, dummy_skill, dummy_terrain, dummy_dayleave, dummy_quarter]
for features in one_hot_encoded:
  day_hikes = pd.merge(left = day_hikes, right = features, left_index = True, right_index = True)

day_hikes = day_hikes.drop(columns = ['Trip Type', 'Skill Level', 'Terrain Type', 'Day of week', 'Quarter'])
day_hikes.head(5)
```

```python
day_hikes.to_csv('Day_Hikes_Encoded.csv', index = False)
```

```python
features = day_hikes.drop(['Year', 'Trip Name', 'Region', 'County', 'Tickets Sold', 'Tickets Available',
                                 'Empty Spots', 'Filled?', 'PercentSold', 'Success'], axis = 1)
target = day_hikes['Success']
```

```python
import sklearn
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
top_features_ordered = []
```

```python
for i in range(len(features)-1):
  best_features = SelectKBest(f_classif, k=i)
  k_best_data = best_features.fit_transform(features, target)
  mask = best_features.get_support()
  new_features = features.columns[mask]
  for item in new_features:
    if item not in top_features_ordered:
      top_features_ordered.append(item)
```

```python
count = 0
for feature in top_features_ordered:
  count +=1
  print(count, ': ', feature)
```
