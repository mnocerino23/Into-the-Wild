# 2021-2022 ITW Day Trips Year in Review

ITW Logo.JPG

Read in dataset containing information and sign up data for all of Into the Wild's day hikes during the 2021-2022 school year at Santa Clara University.

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
day_hikes = pd.read_csv('/content/drive/MyDrive/Data_Science_Projects/Into_the_Wild/Into the Wild Trip Results 2021 - Day Trips.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
print(day_hikes.columns)
```

```python
#Add in feature that represents the  

day_hikes = day_hikes.assign(PercentSold=lambda x: (x['Tickets Sold'] / x['Tickets Available']))
```

```python
day_hikes.head(5)
```

```python
day_hikes.describe()
```

# Recap of Into the Wild 2021 Trips - Trip Type, Skill Level, and Terrain

```python
sns.histplot(data=day_hikes, x="Trip Type", color = "orange")
plt.title('2021-2022 Day Trip Types', fontsize=14)
```

```python
sns.histplot(data = day_hikes, x = "Skill Level", color = "blue")
plt.title('2021-2022 Day Hike Skill Levels', fontsize=14)
```

```python
sns.histplot(data = day_hikes, x = "Terrain Type", color = "magenta")
plt.title('2021-2022 Day Hike Terrain', fontsize=14)
```

```python
sns.histplot(data = day_hikes, x = "Quarter", color = "brown")
plt.title('2021-2022 Day Hikes by Quarter', fontsize=14)
```

# Where we got Wild:

```python
sns.set(rc = {'figure.figsize':(25,8)})
sns.histplot(data = day_hikes, x = "Region",  bins = 20)
plt.title('2021-2022 Day Hikes by Region', fontsize=25)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.histplot(data = day_hikes, x = "County",  bins = 20)
plt.title('2021-2022 Day Hikes by County', fontsize=25)
```

# Trip Distances

```python
sns.histplot(data = day_hikes, x = "Distance (miles)", bins = 5, kde = True, color = 'Green')
plt.title('Driving Distances of 2021-2022 ITW Day Hikes', fontsize=16)
```

Violin Plot using Seaborn:

```python
plt.title('Distribution of Distances of 2021-2022 ITW Day Hikes', fontsize=20)
sns.set_theme(style="whitegrid")
ax = sns.violinplot(x=day_hikes["Distance (miles)"])
```

# Pricing of 2021-2022 Overnights

```python
sns.histplot(data = day_hikes, x = "Price", bins = 10, kde = True, color = 'Purple')
plt.title('Histogram of 2021-2022 ITW Day Hikes Prices', fontsize=20)
plt.xlabel('Trip Price (USD)')
```

Violin Plot using seaborn:

```python
plt.title('Price Distribution of 2021-2022 ITW Day Hikes', fontsize=20)
sns.set_theme(style="whitegrid")
ax = sns.violinplot(x=day_hikes["Price"])
```

# Investigating the lack of sign ups during the 2021-2022 school year

```python
sns.histplot(data = day_hikes, x = "PercentSold", bins = 5, kde = True)
plt.title('Percent Full 2021-2022 Day Hikes', fontsize=20)
plt.xlabel('Trip Price (USD)')
```
